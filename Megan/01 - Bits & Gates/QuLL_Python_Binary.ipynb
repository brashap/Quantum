## Functions

A function allows programmers to implement a particular task once, then call the function to perform that task whenever needed. It can take inputs and return an output, or it can just perform some action.

For example, the following function returns the square of its input:

```python
def square(x):
  """
  Input:
    x: a number

  Output:
    x squared
  """

  return x**2
```

```python
square(5)
```

Functions are really helpful when we want to repeat the same task multiple times, but with different inputs. For example, suppose you want to write your name in binary. You could do this with the help of two functions:
1.   a function that converts a letter into an integer
2.   the built-in "bin" function that converts an integer into binary

The below code implements the first function. Don't worry about understanding it in detail.

```python
def letter2int(char):
  """
  Input:
    char: a letter

  Output:
    the corresponding integer
  """

  alphabet = 'abcdefghijklmnopqrstuvxyz '
  return alphabet.index(char.lower()) + 1
```

The below shows an example of how the _letter2int_ function can be used to encode a letter into a number.

```python
letter2int('e')
```

Now we can use both functions together to encode letters into binary!

```python
x = letter2int('e')
bin(x)
```

We can even do the conversion in a single line by directly using the output of "letter2int" as the input to "bin".

```python
bin( letter2int('e') )
```

The buit-in print function is useful for printing out multiple lines of output.

```python
print( bin( letter2int('Q') ) )
print( bin( letter2int('C') ) )
print( bin( letter2int('a') ) )
print( bin( letter2int('M') ) )
print( bin( letter2int('P') ) )
```

## Loops

You might have thought it was annoying to keep repeating mostly the same code to convert each letter of your name one by one. Luckily, we can use a loop to iterate over the letters in your name and automatically repeat the same code for each letter.

For example, the below "for loop" prints out each letter of QCaMP.

```python
word = 'QCaMP'

for letter in word:
  print(letter)
```
